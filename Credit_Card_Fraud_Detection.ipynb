Build a sample dataset containing similar distribution of normal transactions and fraud transactions

```python
legit_sample = legit.sample(n=492)
```

```python
new_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby("Class").mean()
```

```python
# Splitting the data into features and target
X = new_df.drop(columns = 'Class', axis = 1) 
Y =new_df['Class']
```

```python
print(X)
print(Y)
```

```python
# Split the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model trainig using Logistic Regression
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# Model Evaluation and accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data:',training_data_accuracy)
```

```python
# Model Evaluation and accuracy score on testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on testing data:', testing_data_accuracy)
```

```python
Y_pred = model.predict(X_test)
cm = confusion_matrix(Y_test, Y_pred)
cm
```

```python
#Confusion matrix

sb.heatmap(cm, annot=True, cbar=None, fmt = 'g')
plt.title("Confusion Matrix MLP"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```
